```python
import numpy as np
import matplotlib.pyplot as plt

def square_wave(frequency, duration, sample_rate, harmonics=10):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    wave = np.zeros_like(t)
    
    for n in range(1, harmonics * 2, 2):  # Harmoniques impaires
        wave += (1 / n) * np.sin(2 * np.pi * n * frequency * t)
    
    wave *= 4 / np.pi  # Normalisation
    return t, wave

# Paramètres
frequency = 440  # Fréquence en Hz (La4)
duration = 1.0   # Durée en secondes
sample_rate = 44100  # Taux d'échantillonnage en Hz

# Générer l'onde carrée
t, wave = square_wave(frequency, duration, sample_rate)

# Visualiser
plt.plot(t[:1000], wave[:1000])  # Afficher seulement 1000 échantillons
plt.title("Onde Carrée (Série de Fourier)")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
